## 결측치 처리

표준편차값이 0인 컬럼들을 삭제

-결측치를 최빈값으로 처리하고 상관관계 분석

-결측치를 중앙값으로 처리하고 상관관계 분석

```python
# summary_statistics = train_amode.describe()

# std_0 = summary_statistics.columns[(summary_statistics.loc['std'] == 0) | (summary_statistics.loc['std'] == 0.0)]
# print(std_0)
```

```python
#최빈값으로 결측치 채우기
train_amode = train_a.iloc[:,6:].copy()
for col in train_a.columns:
    mode_value = train_a[col].mode()[0]  # 최빈값 계산
    train_amode[col] = train_a[col].fillna(mode_value)
# train_amode.X_135.unique()
```

#### 동일한 값만 들어가 표준 편차 0인 컬럼 삭제

```python
std_00 = []
for i in train_amode.columns:
    if len(train_amode[i].unique()) == 1:
        print(i)
        std_00.append(i)
        
train_amode.drop(columns = std_00, inplace = True)
train_amode
```

```python
#train_amode.drop(columns = ['PRODUCT_ID', 'Y_Quality', 'TIMESTAMP'], inplace = True)
train_amode.to_csv('./data/train_amode.csv', index = False)
```

```python
# 중앙값으로 결측치 처리
train_amedi = train_a.iloc[:,6:].copy()
for col in train_amedi.columns:
    medi_value = train_amedi[col].median()  # 중앙값 계산
    train_amedi[col] = train_amedi[col].fillna(medi_value) 
```

```python
#train_amedi.drop(columns = ['PRODUCT_ID', 'Y_Quality', 'TIMESTAMP'], inplace = True)
train_amedi.to_csv('./data/train_amedi.csv', index = False)
```

### 상관관계분석(히트맵)으로 중요한 컬럼만 추출한다.

```python
df = train_amode.copy()
```

```python
df.drop(columns = ['PRODUCT_ID', 'Y_Quality', 'TIMESTAMP', 'Y_Class'], inplace = True)
```

```python
df
```

```python
corr_matrix = df.corr()
```

```python
# Y_Class 제외 0.8보다 높은 상관계수를 가진 컬럼 추출 -> 삭제
high_corr_col = set()
for i in range(len(corr_matrix.columns)):
    for j in range(i+1, len(corr_matrix.columns)):
        if abs(corr_matrix.iloc[i, j]) > 0.8:
            high_corr_col.add(corr_matrix.columns[i])
            high_corr_col.add(corr_matrix.columns[j])


print("Selected high correlation columns:", high_corr_col) #1285개 컬럼 삭제
```

```python
# 새로운 DataFrame 생성
df_corr = train_amode.drop(columns = high_corr_col)
len(df_corr)
```

```python
df_corr.shape
```

```python
df_corr.to_csv('./data/train_amode2_corr.csv', index = False)
```

```python
corr_matrix = df.corr()
highcorr_lst = []
for i in range(1, len(corr_matrix.columns)):
    if abs(corr_matrix.iloc[0, i]) > 0.3:
        highcorr_lst.append(corr_matrix.columns[i])

print("Columns in df:", df.columns)
print("Selected high correlation columns:", highcorr_lst)
df_corr = df[highcorr_lst]
```
