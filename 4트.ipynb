```python
def seed_everything(seed):
    random.seed(seed) #파이썬 자체 모듈 random 모듈의 시드 고정
    os.environ['PYTHONHASHSEED'] = str(seed) 
    np.random.seed(seed)

seed_everything(37) # Seed 고정
```

```python
#PRODUCT_CODE 원핫 인코딩하기
#원핫인코딩 -> train set
tpc = pd.get_dummies(train_100['PRODUCT_CODE'])
for i in tpc:
    train_100.insert(1, i, tpc[i])
train_100.drop(columns=['PRODUCT_CODE'], inplace=True)
tpc = pd.get_dummies(train_050['PRODUCT_CODE'])
for i in tpc:
    train_050.insert(1, i, tpc[i])
train_050.drop(columns=['PRODUCT_CODE'], inplace=True)
tpc = pd.get_dummies(train_010['PRODUCT_CODE'])
for i in tpc:
    train_010.insert(1, i, tpc[i])
train_010.drop(columns=['PRODUCT_CODE'], inplace=True)

#원핫인코딩 -> test set
tpc = pd.get_dummies(test_100['PRODUCT_CODE'])
for i in tpc:
    test_100.insert(0, i, tpc[i])
test_100.drop(columns=['PRODUCT_CODE'], inplace=True)
tpc = pd.get_dummies(test_050['PRODUCT_CODE'])
for i in tpc:
    test_050.insert(0, i, tpc[i])
test_050.drop(columns=['PRODUCT_CODE'], inplace=True)
tpc = pd.get_dummies(test_010['PRODUCT_CODE'])
for i in tpc:
    test_010.insert(0, i, tpc[i])
test_010.drop(columns=['PRODUCT_CODE'], inplace=True)
```
